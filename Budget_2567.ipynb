```python
df = pd.read_csv('budget.csv')
df.head()
```

```python
unique_ministries = df['MINISTRY'].unique()
print(unique_ministries)
```

```python
filtered_df = df[df['MINISTRY'] == 'กระทรวงการอุดมศึกษา วิทยาศาสตร์ วิจัยและนวัตกรรม']
column_names = df.columns.tolist()
print(column_names)
```

```python
unique_unit = filtered_df['BUDGETARY_UNIT'].unique()
print(unique_unit)
```

```python
filtered_df1 = filtered_df[filtered_df['BUDGETARY_UNIT'] == 'มหาวิทยาลัยมหิดล']
filtered_df1.head(10)
```

```python
filtered_df1.to_csv('budget_mu.csv', index=False)
```

```python
filtered_df1.to_excel('budget_mu.xlsx', index=False)
```

```python
unique_item = filtered_df1['BUDGET_PLAN'].unique()
print(unique_item)
```

```python
# Remove commas and convert to numeric
filtered_df1['NUMBER_AMOUNT'] = (
    df['AMOUNT']
    .astype(str)  # Ensure the column is treated as strings
    .str.replace(',', '', regex=False)  # Remove commas
    .str.replace(' ', '', regex=True) # remove whitespace
    .apply(pd.to_numeric, errors='coerce')  # Convert to numeric, handle errors
)
```

```python
filtered_df1.head(2)
```

```python
sum = filtered_df1['NUMBER_AMOUNT'].sum()
# Display the sum in scientific notation
print(f"{sum:,}")
print(f"{sum:.2e}")
```

```python
#List of budget plans to filter
target_budget_plans = ['แผนงานบุคลากรภาครัฐ', 'แผนงานพื้นฐานด้านการพัฒนาและเสริมสร้างศักยภาพทรัพยากรมนุษย์', 'แผนงานยุทธศาสตร์พัฒนาศักยภาพคนตลอดช่วงชีวิต', 'แผนงานยุทธศาสตร์เสริมสร้างให้คนมีสุขภาวะที่ดี', 'แผนงานยุทธศาสตร์สร้างความเสมอภาคทางการศึกษา']

# Filter the DataFrame
filtered_df2 = filtered_df1[filtered_df1['BUDGET_PLAN'].isin(target_budget_plans)]
```

```python
sum = filtered_df2['NUMBER_AMOUNT'].sum()
print(f"{sum:,}")
print(f"{sum:.2e}")
```

```python
# Group by 'BUDGET_PLAN' and calculate the sum
grouped_data = filtered_df2.groupby('BUDGET_PLAN')['NUMBER_AMOUNT'].sum()

# Create a new DataFrame
result_df = grouped_data.reset_index(name='TOTAL_AMOUNT')
print(result_df)
result_df.to_excel('budget_mu_group.xlsx', index=False)
```

```python
# Create pile chart
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'TH Sarabun New'  # Or another Thai-compatible font
plt.figure(figsize=(8, 6))
plt.pie(result_df['TOTAL_AMOUNT'], labels=result_df['BUDGET_PLAN'], autopct='%1.1f%%', startangle=140)
plt.title('Budget Plan Distribution')
plt.axis('equal')
plt.show()  # Display the pie chart
```
