Let's calculate the total budget for each unique `plan_name` within the `filtered_df1` DataFrame.

```python
sum = filtered_df1['p_total_bud'].sum()
# Display the sum in scientific notation
print(f"{sum:,}")
print(f"{sum:.2e}")
```

```python
# Group by 'plan_name' and sum 'p_total_bud'
total_budget_per_plan = filtered_df1.groupby('plan_name')['p_total_bud'].sum().reset_index()

# Display the result
print("Total Budget for Each Plan Name:")
display(total_budget_per_plan)
```

Here's a pie chart to visualize the distribution of the total budget for each `plan_name`.

```python
import matplotlib.pyplot as plt

# Ensure the font is set for Thai characters
plt.rcParams['font.family'] = 'TH Sarabun New'

fig, ax = plt.subplots(figsize=(10, 8))

# Define sizes and labels from the total_budget_per_plan DataFrame
sizes = total_budget_per_plan['p_total_bud']
labels = total_budget_per_plan['plan_name']

# Calculate percentages
total = sizes.sum()
percentages = [f'{p/total*100:.1f}%' for p in sizes]

# Create labels for the legend with percentages
legend_labels = [f'{label} ({percentage})' for label, percentage in zip(labels, percentages)]

wedges, texts = ax.pie(sizes, startangle=90, wedgeprops=dict(width=0.3)) # Use wedgeprops for a donut-like chart

ax.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
ax.legend(wedges, legend_labels, title="Total Budget Distribution by Plan Name",
          loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Total Budget Distribution by Plan Name')
plt.show()
```
